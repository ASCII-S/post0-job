# 1.1.2 字符串

```python
# 字符串拼接
# +
print ("hello" + " world")
# join
print(" ".join([str(i) for i in range(10)]))
```

```python
# 格式化字符串
from re import X


name = "Alice"
age = 26
print("%:")
print("hello, is %s" % name) 
print("age is %d" % age)
print("%(name)s is %(age)d years old" % {"name": "Alice", "age": 26})

print("format:")
print("name is {0}, and {0} is {1} years old.".format("Alice", "26"))

print("f-string:")
print(f"name is {name.upper()}, age is {age}, score is {age:.2f}")
```

```python
# 字符串切片
s = "Hello World"

print(s[:5])
print(s[-5:])

print(s[:5:2])
```
